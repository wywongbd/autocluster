```python
import pandas as pd
df = pd.read_csv("Financial_Distress.csv")
```

```python
def make_class(financial_distress):
    if financial_distress > -0.5:
        return 'Healthy'
    else:
        return 'Financially Distressed'
```

```python
df['class'] = df.apply(lambda row: make_class(row['Financial Distress']), axis=1)
```

```python
df.to_csv('Financial_Distress.csv', mode = 'w')
```
